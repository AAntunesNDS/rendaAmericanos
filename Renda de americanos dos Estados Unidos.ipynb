### Carregando e dividindo o dataset em dois dataframes: numericos e categoricos

```python
names = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship',
    'race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']

data = pd.read_csv('adult.data',  na_values='?', sep= ',\s+', names=names)

data = data.drop('education', axis=1)

dataCategorias = data.select_dtypes(['object']).columns
```

### Transformando nulos na moda

```python
for i in ['workclass', 'occupation', 'native-country']:
    data[i] = data[i].replace(np.nan, data[i].mode()[0])
```

### Binarizando categorias

```python
data = pd.get_dummies(data, columns=dataCategorias, prefix=dataCategorias, drop_first=True)
data.head()
```

### Separando features e label

```python
renda = data['income_>50K']
renda.head()
```

```python
features = data.drop('income_>50K', axis = 1)
features.head()
```

```python
renda.shape, features.shape
```

### Modelos

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron

y = renda
X = features

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=3) 
```

```python
from sklearn.linear_model import Perceptron

classifiers = [
    Perceptron()]

for clf in classifiers:
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    print(format(clf.predict([np.array(data.head())[0][:82]])))
    print(m.accuracy(y_pred, y_test) , "\n")

print(y_pred[0:10])
```
